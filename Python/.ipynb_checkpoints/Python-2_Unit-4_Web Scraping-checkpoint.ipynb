```python
def scrape_mblock(movie_block):
    
    movieb_data ={}
  
    try:
        movieb_data['name'] = movie_block.find('a').get_text() # Name of the movie
    except:
        movieb_data['name'] = None

    try:    
        movieb_data['year'] = movie_block.find('span',{'class': 'lister-item-year'}).get_text()[1:-1] # Release year
    except:
        movieb_data['year'] = None

    try:
        movieb_data['rating'] = float(movie_block.find('div',{'class':'inline-block ratings-imdb-rating'}).get('data-value')) #rating
    except:
        movieb_data['rating'] = None

    return movieb_data
    
```

```python
def scrape_m_page(movie_blocks):
    
    page_movie_data = []
    num_blocks = len(movie_blocks)
    
    for block in range(num_blocks):
        page_movie_data.append(scrape_mblock(movie_blocks[block]))
    
    return page_movie_data
```

```python
scrape_m_page(movie_blocks)
```

```python
import requests
import bs4
url = 'https://subslikescript.com/movies'
result = requests.get(url).text
soup = bs4.BeautifulSoup(result,'html.parser')
print(soup.prettify())
```

```python
soup.findAll('a',href=True)
```

```python
len(soup.findAll('a',href=True))
```

```python
box=soup.find('article',class_='main-article')
print(box)
```

```python
box.findAll('a',href=True)
```

```python
len(box.findAll('a',href=True))
```

```python
links= []
for link in box.findAll('a',href=True):
    links.append(link['href'])
print(links)
```

```python
root = 'https://subslikescript.com/'
for link in links:
    url = root + link
    result = requests.get(url).text
    soup = bs4.BeautifulSoup(result,'html.parser')
    transcript = soup.find('div',class_='full-script').get_text()
    title = soup.find('h1').text
    try:
        with open (f'E:\Files\{title}.txt','w',encoding='utf-8') as f:
            f.write(transcript)
    except Exception as e:
        print(e)
```

```python
import bs4
soup=bs4.BeautifulSoup(open('politifact_factchecks.html',encoding='utf-8').read())
print(soup.prettify())
```

```python
fact = soup.findAll('li',class_='o-listicle__item')
print(fact)
```

```python
state=[]
for i in fact:
    a = i.find('div', class_='m-statement__quote').text.strip()
    state.append(a)
print(state)
```

```python
author=[]
for i in fact:
    b=i.find('footer',class_='m-statement__footer').text[3:-15].strip()
    print(b)
    author.append(b)
print(author)
```

```python
source=[]
for i in fact:
    c=i.find('div',class_='m-statement__meta').find('a').text.strip()
    source.append(c)
print(source)
```

## Multipage

```python
import bs4
import pandas as pd
import requests
upperframe=[]
for page in range(1,3):
    url='https://www.politifact.com/factchecks/list/?page='+str(page)
    pages=requests.get(url).text
    soup = bs4.BeautifulSoup(pages,'html.parser')
    frame=[]
    links=soup.findAll('li',class_='o-listicle__item')
    for j in links:
        statement=j.find('div',class_='m-statement__quote').text.strip()
        link='https://www.politifact.com'
        link+=j.find('div',class_='m-statement__quote').find('a')['href'].strip()
        Date=j.find('div',class_='m-statement__body').find('footer').text[-15:-1].strip()
        Source=j.find('a',class_='m-statement__name').text.strip()
        frame.append((statement,link,Date,Source))
    upperframe.extend(frame)

data=pd.DataFrame(upperframe,columns=['Statement','Link','Date','Source'])
print(data)
```

```python
data.to_csv('politifact.csv')
```
