# creating a server client connection

```python
%%writefile server_TCP.py

import socket

host=socket.gethostname()
port=2000
Addr=(host,port)

server_socket=socket.socket()
server_socket.bind(Addr)
server_socket.listen()
conn,addr=server_socket.accept()

print("Connected through ",str(addr))

while True:
    data=conn.recv(1024).decode()
    if not data:
        break
    print(data)
    
    data_send=input('-->')
    conn.send(data_send.encode())

    
conn.close()
```

```python
%%writefile client_TCP.py
import socket

host=socket.gethostname()
port=2000
Addr=(host,port)

client_socket=socket.socket()
client_socket.connect((host,port))
message=input('-->')

while message!='':
    client_socket.send(message.encode())
    data=client_socket.recv(1024).decode()
    print('received from server:',data)
    message=input('-->')

client_socket.close()    
```

# Square of a given number:

```python
%%writefile serversquare_tcp.py
import socket

host=socket.gethostname()
port=3000
Addr=(host,port)

server_socket=socket.socket()
server_socket.bind(Addr)
server_socket.listen()
conn,addr=server_socket.accept()

print("Connected through ",str(addr))

while True:
    data=conn.recv(1024).decode()
    if not data:
        break
        
    data1=int(data)*int(data)
    conn.send(str(data1).encode())

    
conn.close()
```

```python
%%writefile clientsquare_TCP.py
import socket

host=socket.gethostname()
port=3000
Addr=(host,port)

client_socket=socket.socket()
client_socket.connect((host,port))
message=input('enter number u want to square of:')

while message!='':
    client_socket.send(message.encode())
    data=client_socket.recv(1024).decode()
    print('received from server:',data)
    message=input('enter number u want to square of:')

client_socket.close() 
```

# UDP Server 

```python
%%writefile udp_server.py
import socket

host=socket.gethostname()
port=3000
Addr=(host,port)

udp_server=socket.socket(type=socket.SOCK_DGRAM)
udp_server.bind(Addr)

while True:
    print('waiting for message')
    data,addr=udp_server.recvfrom(1024)
    if not data:
        break
    print('received',data.decode())
    msg=input('->')
    udp_server.sendto(msg.encode(),addr)
    
udp_server.close()    
```

```python
%%writefile udp_client.py
import socket

host=socket.gethostname()
port=3000
Addr=(host,port)

udp_client=socket.socket(type=socket.SOCK_DGRAM)

while True:
    data=input('Enter data:')
    if not data:
        break
        
    udp_client.sendto(data.encode(),Addr)
    print('ready to receive')
    
    data1,addr=udp_client.recvfrom(1024)
    if not data:
        break
    
    print('received',data1.decode())

udp_client.close()    
```

# Simple web-browser in python:

```python
import socket

mysock=socket.socket()
mysock.connect(('www.ljku.edu.in',80))
cmd='GET https:www.ljku.edu.in/lju-at-a-glance HTTP/1.0\r\n\r\n'.encode()
mysock.send(cmd)

while True:
    data=mysock.recv(1024)
    if len(data)<1:
        break
    print(data.decode(),end='')
    
mysock.close()    
```

# Simple HTTP Server in python:

```python
from socket import *

def createserver():
    serversocket=socket()
    addr=('localhost',9000)
    serversocket.bind(addr)
    serversocket.listen()
    
    while True:
        (clientsocket,ar)=serversocket.accept()
        r=clientsocket.recv(1024).decode()
        print(r)
        
        data='HTTP/1.1 200 ok\r\n'
        data+='content-type:text/html; charset=utf-8\r\n'
        data+='\r\n'
        data+='<html><body>Hello World <a href="https://www.wikipedia.org">Waah</a></body></html>\r\n\r\n'
        
        clientsocket.send(data.encode())
        
print("Access http://localhost:9000")
createserver()
```

```python
#Current Wheather
import requests,json

def get_coordinates():
    api_key=input("Enter your API key:")
    city=input('Enter the city name:')
    
    request_url="https://api.openweathermap.org/geo/1.0/direct?q="+city+"&appid="+api_key
    response=requests.get(request_url)
    coordinates=response.json()
    
    lat=coordinates[0]['lat']
    lon=coordinates[0]['lon']
    
    req="https://api.openweathermap.org/data/2.5/weather?lat="+str(lat)+"&lon="+str(lon)+"&appid="+api_key
    res=requests.get(req)
    coordinates=res.json()
#     print(json.dumps(coordinates,indent=5))
    
    
#     to display specific descriptions:
    desc=coordinates['weather'][0]['description']
    temp=coordinates['main']['temp']
    pre=coordinates['main']['pressure']
    hum=coordinates['main']['humidity']
    speed=coordinates['wind']['speed']
    visi=coordinates['visibility']
    return desc,temp,pre,hum,speed,visi

get_coordinates()    
```

# Generate wheather forecast for next 5 days in city:

```python
import requests,json

quality_indices={1:'good',2:'fair',3:'moderate',4:'poor',5:'very poor'}

def get_coordinates():
    api_key=input("Enter your API key:")
    city=input('Enter the city name:')
    
    request_url="https://api.openweathermap.org/geo/1.0/direct?q="+city+"&appid="+api_key
    response=requests.get(request_url)
    coordinates=response.json()
    
    lat=coordinates[0]['lat']
    lon=coordinates[0]['lon']
    
    req="https://api.openweathermap.org/data/2.5/forecast?lat="+str(lat)+"&lon="+str(lon)+"&appid="+api_key
    res=requests.get(req)
    coordinates=res.json()
#     print(json.dumps(coordinates,indent=5))

    d={'data_time':[],'temp':[],'press':[],'humi':[],'wea':[],}
    
    for i in coordinates['list']:
        d['data_time'].append(i['dt_txt'])
        d['temp'].append(i['main']['temp'])
        d['press'].append(i['main']['pressure'])
        d['humi'].append(i['main']['humidity'])
        d['wea'].append(i['weather'][0]['description'])
    return d

d=get_coordinates()
```

```python
import pandas as pd
df=pd.DataFrame(d)
print(df)
```

```python
# to create csv
df.to_csv("forecast.csv")
```

```python
import requests
from bs4 import BeautifulSoup

root='https://subslikescript.com/'
for link in links:
    
    url=root+link
    result=requests.get(url)
    content=result.text
    sopu=BeautifulSoup(content,'html.parser')
    tra=sopu.find('div',class_='full-script').text.strip()
    title=sopu.find('h1').text.strip()
    try:
        with open(f"{title}1.txt","w",encoding='utf-8') as f:
            f.write(tra)
    except Exception as e:
        print(e)  
```

# Fact Check:-

```python
import requests
from bs4 import BeautifulSoup
url="https://www.politifact.com/factchecks/"
result=requests.get(url)
content=result.text
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
fact=soup.find_all('li',class_="o-listicle__item")
print(fact)
```

```python
fa=soup.find_all("li",class_="o-listicle__item")
#print(fa)
statement=[]
for i in fact:
    a=i.find('div',class_='m-statement__quote').text.strip()
    statement.append(a)
print(statement)
```

```python
fa=soup.find_all("li",class_="o-listicle__item")
#print(fa)
statement=[]
for i in fact:
    a=i.find('a',class_='m-statement__name').text.strip()
    statement.append(a)
print(statement)
```

```python
fa=soup.find_all("li",class_="o-listicle__item")
#print(fa)
dates=[]
for i in fact:
    a=i.find('div',class_='m-statement__body').find("footer").text[-15:-1].strip()
    dates.append(a)
print(dates)
```
