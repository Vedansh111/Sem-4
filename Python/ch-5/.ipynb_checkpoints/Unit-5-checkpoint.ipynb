* Machine Learning:-

-> Uses statical techiniques to give computer system the activity to learn with data, without being explicity programmed.
-> Art of programming computer so that they can learn from data.

> Traditional Learning:-

- data,program->computer->output

> Machine Learning:-

- data,output->computer->program


: Supervised Learning - algorithm includes desired solution directly, called labels(labeled data)
: Unsupervised Learning - algorithm includes desired solution indirectly, called unlabel(unlabeled data)
: Reinformence Learning - learning or our own thourgh trial and error

```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as m

df=pd.read_csv('Book1.csv')
df.head()
```

```python
x=df.iloc[:,0:1]
print(x)
type(x)
```

```python
y=df.iloc[:,-1]
print(y)
type(y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
d=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
print(d)
```

```python
m.scatter(df['cgpa'],df['package'])
m.plot(x_test,y_pred,color='red')
m.show()
print(lr.coef_)#m
print(lr.intercept_)#c
```

```python
from sklearn import metrics

print('MAE',metrics.mean_absolute_error(y_test,y_pred))
print('MSE',metrics.mean_squared_error(y_test,y_pred))
```

```python
y_pred=lr.predict([[5.5,45000,0,10,0,1,0,1]])
print(y_pred)
```

```python
import pandas as pd
data=pd.read_csv('Advertising.csv')
data.head()
```

```python
data.isna().sum()
```

```python
x=data[['TV','Radio','Newspaper']]
# y=data[['Sales']] #will provide dataframe
y=data['Sales']
print(x.shape)
print(type(x))
```

```python
print(y.shape)
print(type(y))
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
diff=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
print(diff)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
from sklearn import metrics

print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:',metrics.mean_squared_error(y_test,y_pred))
print('R2 Score:',metrics.r2_score(y_test,y_pred))
```

```python
import pandas as pd

df=pd.read_csv('insurance.csv')
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
df.isna().sum()
```

```python
# Creating a pairwise plot in Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df)
plt.show()
```

```python
# Plotting a pairplot of your DataFrame
sns.pairplot(df, hue='smoker_yes')
plt.show()
```

```python
# Plotting a scatterplot of the relationship of age and charges
sns.relplot(data=df, x='age', y='charges', hue='smoker_yes')
plt.show()
```

```python
df=pd.get_dummies(data=df,drop_first=True)
print(df)
```

```python
x=df[['age','bmi','smoker_yes']]
y=df['charges']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
diff=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
print(diff)
```

```python
from sklearn import metrics

print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:',metrics.mean_squared_error(y_test,y_pred))
print('R2 Score:',metrics.r2_score(y_test,y_pred))
```

```python
# car data
import pandas as pd

df=pd.read_csv('car data.csv')
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['Age']=2023-df['Year']
df.drop('Year',axis=1,inplace=True)
df.head()
```

```python
# Car_Name is not required
df.drop(labels='Car_Name',axis=1,inplace=True)
df.head()
```

```python
df=pd.get_dummies(data=df,drop_first=True)
print(df)
```

```python
# x=df[['Age','Present_Price','Transmission_Manual','Fuel_Type_Petrol','Fuel_Type_Diesel','Owner','Seller_Type_Individual','Kms_Driven']
x=df.drop('Selling_Price',axis=1)
y=df['Selling_Price']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
diff=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
# print(diff)
m=lr.coef_
print(m)
c=lr.intercept_
print(c)
```

```python
from sklearn import metrics

print('MAE:',metrics.mean_absolute_error(y_test,y_pred))
print('MSE:',metrics.mean_squared_error(y_test,y_pred))
print('R2 Score:',metrics.r2_score(y_test,y_pred))
```

```python
y_pred=lr.predict([[5.5,45000,0,10,0,1,0,1]])
print(y_pred)
```
